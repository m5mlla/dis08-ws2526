## Comma Code 


```python

def comma_code(items):
    """Wandelt eine Liste in einen String wie 'a, b, and c' um."""
    # Fall 1: leere Liste
    if not items:
        return ''

    # Fall 2: nur ein Element
    if len(items) == 1:
        return str(items[0])

    # Fall 3: mehrere Elemente
    # rest = alle außer letztes, last = letztes Element
    rest = items[:-1]
    last = items[-1]
    return ", ".join(map(str, rest)) + ", and " + str(last)

# Tests (so siehst du, ob es funktioniert)
print(comma_code([]))  
print(comma_code(['apples']))  
print(comma_code(['apples', 'bananas'])) 
print(comma_code(['apples', 'bananas', 'tofu', 'cats'])) 
```

## Coin Flip Streaks

```python

import random

def has_six_streak(flips):
    """True, wenn irgendwo 6 gleiche Werte in Folge vorkommen."""
    streak_len = 1  # aktuelle Serienlänge
    for i in range(1, len(flips)):
        if flips[i] == flips[i - 1]:
            streak_len += 1
            if streak_len >= 6:
                return True
        else:
            streak_len = 1
    return False

numberOfStreaks = 0
experiments = 10000
flips_per_experiment = 100

for _ in range(experiments):
    # 100 zufällige H/T
    flips = ['H' if random.randint(0, 1) == 0 else 'T' for _ in range(flips_per_experiment)]
    # Prüfen, ob es eine 6er-Streak gibt
    if has_six_streak(flips):
        numberOfStreaks += 1

chance_percent = numberOfStreaks / (experiments / 100)
print('Chance of streak: %s%%' % chance_percent)
```

## Fantasy Game Inventory

```python

stuff = {'rope': 1, 'torch': 6, 'gold coin': 42, 'dagger': 1, 'arrow': 12}

def displayInventory(inventory):
    print("Inventory:")
    item_total = 0
    for item_name, count in inventory.items():
        print(str(count) + " " + item_name)
        item_total += int(count)
    print("Total number of items: " + str(item_total))

# Test
displayInventory(stuff)
```

## List to Dictionary Function 

```python

def addToInventory(inventory, addedItems):
    for item in addedItems:
        if item in inventory:
            inventory[item] += 1
        else:
            inventory[item] = 1
    return inventory

inv = {'gold coin': 42, 'rope': 1}
dragonLoot = ['gold coin', 'dagger', 'gold coin', 'gold coin', 'ruby']
inv = addToInventory(inv, dragonLoot)
displayInventory(inv)
```
